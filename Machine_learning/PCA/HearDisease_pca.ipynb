```python
df[df.RestingBP>(df.RestingBP.mean()+3*df.RestingBP.std())]
```

```python
df3=df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3.head()
```

```python
df.ChestPainType.unique()
```

```python
df.ExerciseAngina.unique()
```

```python
df.ST_Slope.unique()
```

```python
df4=df3.copy()
```

```python
df4.ExerciseAngina.replace(
    {
        'N':0,
        'Y':1
    },
    inplace=True)
df4.ST_Slope.replace(
    {
        'Down':1,
        'Flat':2,
        'Up':3
    },
    inplace=True)
df4.RestingECG.replace(
    {
        'Nomal':1,
        'ST':2,
        'LVH':3
    },
    inplace=True)
```

```python
df4.head()
```

```python
df5=pd.get_dummies(df4,drop_first=True)
df5.head()
```

```python
X=df5.drop("HeartDisease",axis="columns")
Y=df5.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y,test_size=0.2,random_state=30)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf=RandomForestClassifier()
model_rf.fit(X_train,Y_train)
```

```python
model_rf.score(X_test,Y_test)
```

Use pca to reduce dimesions

```python
X
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(0.95)
X_pca=pca.fit_transform(X)
X_pca
```

```python
X_train_pca,X_test_pca,Y_train,Y_test=train_test_split(X_pca,Y,test_size=0.3)
```

```python
model_rf1=RandomForestClassifier()
```

```python
model_rf1.fit(X_train_pca,Y_train)
```

```python
model_rf1.score(X_test_pca,Y_test)
```

```python
import pickle
```

```python
with open("Trained_heart_pca","wb") as f:
    pickle.dump(model_rf1,f)
```
