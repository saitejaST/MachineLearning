```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
df=pd.DataFrame(digits.data,digits.target)
df.head()
```

```python
df['Target']=digits.target
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df.drop('Target',axis='columns'),df.Target,test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel='rbf')   #we have used radial basic function kernel to find non linear classifier line
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
lmodel=SVC(kernel='linear')   #here we have used linear to make the dataset linearly seperable from the classes
```

```python
lmodel.fit(X_train,Y_train)
```

```python
lmodel.score(X_test,Y_test)
```
