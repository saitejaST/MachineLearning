```python
# now we will be performing the oeprations on digit dataset
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
df=pd.DataFrame(digits.data,digits.target)
df.head()
```

```python
digits.target
```

```python
digits.data
```

```python
dir(digits.data)
```

```python
df['target']=digits.target
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df.drop('target',axis='columns'),df.target,test_size=0.3,random_state=10)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# now we will implement the knnclassifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,Y_train)
```

```python
knn.score(X_test,Y_test)
```

```python
#now we will try by various methods such as confusion matrix
```

```python
y_pred=knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_pred,Y_test)
cm
```

```python
#now we will create a heatmap based on the follwoing data
```

```python
%matplotlib inline
```

```python
import seaborn as sb
```

```python
plt.figure(figsize=(7,5))
sb.heatmap(cm,annot=True)
plt.xlabel("predicted values")
plt.ylabel("Truth values")
```

```python
#now we will evalvuate the performance of the model by using classification report metric
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred,Y_test))
```
